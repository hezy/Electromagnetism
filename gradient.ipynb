```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sx = np.linspace(-1.2, 1.2, 24)
sy = np.linspace(-1, 1, 20)
x, y = np.meshgrid(sx,sy)
```

Electric potential of dipole

$V(\vec r) = \frac{k q}{\left| \vec r \right|} $

```python
# Parameters
k = 9E9
q = 1E-9

# Potential
V = k*q/np.sqrt(x**2 + y**2)
```

```python
#plt.style.use(['nature','notebook'])
plt.contourf(x,y,V, levels=300, cmap='seismic', vmin=-100, vmax=100)
plt.colorbar(label='Electrical Potential [$V$]')
plt.title('Electrical Potential of point-like charge')
plt.xlabel('Horizontal Position [m]')
plt.ylabel('Vertical Position [m]')
```

Electric field of the same dipole

$ \vec E = k \frac{q}{\left| (\vec r \right|^{3}} \vec r $

```python
Ex = k*q*x/(x**2 + y**2)**(3/2)
     
Ey = k*q*y/(x**2 + y**2)**(3/2)
```

```python
E = np.sqrt(Ex**2 + Ey**2)
cos = Ex/E
sin = Ey/E

E_max = 400
E[E>E_max] = E_max
E[E<-E_max] = -E_max
```

```python
#plt.style.use(['nature','notebook'])
plt.quiver(x, y, cos, sin, E, pivot='mid')
plt.colorbar(label='Electrical Field [$V/m$]')
plt.title('Electrical Field of point-like charge')
plt.xlabel('Horizontal Position [m]')
plt.ylabel('Vertical Position [m]')
```

```python
dx = x[0,1] - x[0,0]
dy = y[1,0] - y[0,0]

Ex_, Ey_ = np.gradient(V, dx, dy)
Ex_ = -1 * Ex_
Ey_ = -1 * Ey_

E_ = np.sqrt(Ex_**2 + Ey_**2)
cos_ = Ex_/E_
sin_ = Ey_/E_

E_max_ = 400
E_[E_>E_max_] = E_max_
E_[E_<-E_max_] = -E_max_
```

```python
plt.contourf(x,y,Ex_, levels=300, cmap='seismic')
plt.colorbar(label='Ex [$V/m$]')
plt.title('-dV/dx')
```

```python
plt.contourf(x,y,Ey_, levels=300, cmap='seismic')
plt.colorbar(label='Ey [$V/m$]')
plt.title('-dV/dy')
```

```python
plt.quiver(x, y, sin_, cos_, E_, pivot='mid')
plt.colorbar(label='Electrical Field [$V/m$]')
plt.title('- grad V')
plt.xlabel('Horizontal Position [m]')
plt.ylabel('Vertical Position [m]')
plt.show()
```
