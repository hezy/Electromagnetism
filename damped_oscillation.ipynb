```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
from scipy.integrate import odeint
from scipy.integrate import solve_ivp
```

Newton 2nd law for damped oscillator:

$ -kx -bv = ma $

To use scipy's odeint, we need to write a 1st ordered ODE, so we'll use:

$ \frac{dv}{dt} = -\frac{k}{m}x -\frac{b}{m}v $

$ \frac{dx}{dt} = v $

and put it in vector form:

$ \vec S = (x , v) $

$ \frac{d \vec S}{dt} = A \vec S $

```python
def dSdt(S, t, k, b, m):
    x, v = S
    return [v,                     # dx/dt
            -k/m * x - b/m * v]    # dv/dt
```

```python
k = 400
b = 1.5
m = 0.1
x_0 = 0.5
v_0 = 0
S_0 = (x_0, v_0)
t = np.linspace(0, 1, 200)
```

```python
S = odeint(dSdt, S_0, t=t, args=(k, b, m))
```

```python
x = S.T[0]
v = S.T[1]
fig, axes = plt.subplots(2, 1, figsize=(8,8))

ax1 = axes[0]
ax1.plot(t, x, '-', color='r')
ax1.set_xlabel('time(s)')
ax1.set_ylabel('position(m/s)')

ax2 = axes[1]
ax2.plot(t, v, '-', color='b')
ax2.set_xlabel('time(s)')
ax2.set_ylabel('velocity(m/s)')
#ax2.legend()

fig.tight_layout()
fig.suptitle('damped oscilations', y=1.03, fontsize=25)
plt.show()
```
