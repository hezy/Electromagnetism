```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
from scipy.integrate import odeint
from scipy.integrate import solve_ivp
```

$ -kx -bv = ma $

$ \frac{dv}{dt} = -\frac{k}{m}x -\frac{b}{m}v $

$ \frac{dx}{dt} = v $

$ \vec S = (x , v) $

$ \frac{d \vec S}{dt} = A \vec S $

```python
def dSdt(S, t, k, b, m):
    x, v = S
    return [v,                     # dx/dt
            -k/m * x - b/m * v]    # dv/dt
```

```python
k = 100
b = 2
m = 0.1
x_0 = 0.2
v_0 = 0
S_0 = (x_0, v_0)
t = np.linspace(0, 1, 100)
print(S_0)
```

```python
S = odeint(dSdt, S_0, t=t, args=(k, b, m))
```

```python
#print(S)
x = S.T[0]
v = S.T[1]
plt.plot(t, x)
plt.plot(t, v)
```
