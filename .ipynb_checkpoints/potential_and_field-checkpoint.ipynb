```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
s = np.arange(-2, 2, 0.1)
x, y = np.meshgrid(s,s)
```

```python
f = x * np.exp(-(x**2 + y**2))
```

```python
plt.style.use(['notebook'])
plt.contour(x,y,f, cmap='seismic')
plt.colorbar(label='Electrical Potential [$V$]')
plt.title(r'2D Scalar function f=e^{-(x^2+y^2)}')
plt.xlabel('x')
plt.ylabel('y')
```
