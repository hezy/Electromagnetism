```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
_ = np.linspace(-1, 1, 200)
x, y = np.meshgrid(_,_)
```

```python
k = 9E9
q1 = 1E-9
x1, y1  = -0.3, 0
q2 = 1E-9
x2, y2  = +0.3, 0
V = k*q1/np.sqrt((x-x1)**2+(y-y1)**2) - k*q2/np.sqrt((x-x2)**2+(y-y2)**2)
```

```python
plt.style.use(['default'])
plt.contourf(x,y,V, levels=300, cmap='seismic', vmin=-300, vmax=300)
plt.colorbar(label='Electrical Potential [$Volt$]')
plt.xlabel('Horizontal Position [m]')
plt.ylabel('Vertical Position [m]')
```
