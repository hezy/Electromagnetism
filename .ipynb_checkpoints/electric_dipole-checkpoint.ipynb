```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
_ = np.linspace(-1, 1, 100)
x, y = np.meshgrid(_,_)
```

$V(\vec r) = \frac{k q_1}{\left| \vec r - \vec r_1 \right|} + \frac{k q_2}{\left| \vec r - \vec r_2 \right|}$

```python
k = 9E9
q1 = 1E-9
x1, y1  = -0.3, 0
q2 = -1E-9
x2, y2  = 0.3, 0
V = k*q1/np.sqrt((x-x1)**2+(y-y1)**2) + k*q2/np.sqrt((x-x2)**2+(y-y2)**2)
```

```python
#plt.style.use(['nature','notebook'])
plt.contourf(x,y,V, levels=300, cmap='seismic', vmin=-100, vmax=100)
plt.colorbar(label='Electrical Potential [$V$]')
plt.title('Electrical Potential of a Dipole')
plt.xlabel('Horizontal Position [m]')
plt.ylabel('Vertical Position [m]')
```

plt.hist(V.ravel(), bins=100, histtype='step',label='V')
plt.legend()
plt.xlabel('Electric Potential Magnitude')
plt.ylabel('Frequency')
plt.show()

```python
_ = np.linspace(-1, 1, 24)
x, y = np.meshgrid(_,_)
```

$ \vec E = k \frac{q_1}{\left| (\vec r - \vec r_1 \right|^{3}} (\vec r - \vec r_1) + k \frac{q_2}{\left| (\vec r - \vec r_2 \right|^{3}} (\vec r - \vec r_2) $

```python
Ex = k*q1*(x-x1)/((x-x1)**2+(y-y1)**2)**(3/2) + k*q2*(x-x2)/((x-x2)**2+(y-y2)**2)**(3/2)
Ey = k*q1*(y-y1)/((x-x1)**2+(y-y1)**2)**(3/2) + k*q2*(y-y2)/((x-x2)**2+(y-y2)**2)**(3/2)
```

```python
E = np.sqrt(Ex**2 + Ey**2)
cos = Ex/E
sin = Ey/E

E_max = 400
E[E>E_max] = E_max
E[E<-E_max] = -E_max
```

```python
#plt.style.use(['nature','notebook'])
plt.quiver(x, y, cos, sin, E, pivot='mid')
plt.colorbar(label='Electrical Field [$V/m$]')
plt.xlabel('Horizontal Position [m]')
plt.ylabel('Vertical Position [m]')
```

```python
plt.hist(E.ravel(), bins=100, histtype='step',label='E')
plt.legend()
plt.xlabel('Electric Field Magnitude')
plt.ylabel('Frequency')
plt.show()
```

```python
dx = x[0,1] - x[0,0]
dy = y[1,0] - y[0,0]
#Ex = -1 * np.gradient(V, dx, dy)[0]
#Ey = -1 * np.gradient(V, dx, dy)[1]
Ex, Ey = np.gradient(V, dx, dy)
```

```python
plt.contourf(x,y,Ex, levels=300, cmap='seismic')
plt.colorbar(label='Ex [$V/m$]')
```

```python
plt.contourf(x,y,Ey, levels=300, cmap='seismic')
plt.colorbar(label='Ey [$V/m$]')
```
